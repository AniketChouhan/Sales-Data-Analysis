## Generating New Features 

###### Extracting Week, Day and Year from Order Date

```python
Sales['Week_Day'] = Sales['Order Date'].dt.day_name()
Sales['Day'] = Sales['Order Date'].dt.day
Sales['Year'] = Sales['Order Date'].dt.year
```

```python
# Mapping of month numbers to month names
month_names = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}

# Add a new column "Month_Name" based on the "Month" column
Sales['Month_Name'] = Sales['Month'].map(month_names)
```

```python
Sales['Time_Frame']=''

for index, row in Sales.iterrows():
    hour = row['Hour']
    if 0<= hour <5:
        Sales.at[index, 'Time_Frame'] = 'Mid Night'
    elif 5<= hour <12:
        Sales.at[index, 'Time_Frame'] = 'Morning'
    elif 12<= hour <16:
        Sales.at[index, 'Time_Frame'] = 'Afternoon'
    elif 16<= hour <20:
        Sales.at[index, 'Time_Frame'] = 'Evening'
    else:
        Sales.at[index, 'Time_Frame'] = 'Night'
```

```python
Sales['Month_Segment'] = pd.cut(Sales['Order Date'].dt.day, bins=[0, 10, 20, 32], 
                             labels=['Start of Month', 'Middle of Month', 'End of Month'], 
                             right=False)
```

```python
Drop = Sales.drop(['Month'], axis=1, inplace = True)
```

```python
# Get the list of column names except for non-numeric columns
numeric_columns = Sales.select_dtypes(include=['int64', 'float64']).columns

# Define the number of columns per row in subplots
columns_per_row = 2

# Calculate the number of rows needed
num_columns = len(numeric_columns)
num_rows = (num_columns + columns_per_row - 1) // columns_per_row

# Create subplots
fig, axes = plt.subplots(num_rows, columns_per_row, figsize=(12, 8))

# Flatten the axes array for easier indexing
axes = axes.ravel()

# Iterate through each numeric column and create a violin plot
for i, column in enumerate(numeric_columns):
    sns.violinplot(y=Sales[column], ax=axes[i], color='skyblue')
    axes[i].set_title(f'Violin Plot of {column}', fontsize=12)
    axes[i].set_ylabel(column)
    axes[i].set_xlabel('Density')
    axes[i].grid(True)

# Adjust layout
plt.tight_layout()
plt.subplots_adjust(hspace=0.5)
plt.show()
```
